# Data Preprocessing 


## Import the libraries

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from matplotlib import rc
import seaborn as sns
#Accuracy packages
from sklearn.metrics import mean_squared_error
from sklearn import metrics
```

## Import the dataset

```python
sample_submission= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/SampleSubmission.csv')
sample_submission.columns= [col.replace(' ', '_') for col in sample_submission.columns]
sample_submission.head()
```

```python
sample_submission.shape
```

## Test dataset

```python
test_set= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Test.csv')
test_set.columns= [col.replace(' ', '_') for col in test_set.columns]
test_set.head()
```

## Train dataset

```python
#formating train dataframe 
train_set= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Train.csv')
train_set.columns= [col.replace(' ', '_') for col in train_set.columns]
train_set.head()
```

## Rider dataset

```python
rider= pd.read_csv('https://raw.githubusercontent.com/thembeks/Regression-Sendy-Logistics-Challenge-Team-14/Predict/Riders.csv')
rider.columns= [col.replace(' ', '_') for col in rider.columns]
rider.head()
```

## Merge Rider and Train/Test dataset into a new dataset

```python
Train = pd.merge(train_set, rider, how ='left', left_on=['Rider_Id'], right_on =['Rider_Id'])
Train.head()
```

```python
Test = pd.merge(test_set, rider, how ='left', left_on=['Rider_Id'], right_on =['Rider_Id'])
Test.head()
```

## Data cleanup

```python
#Converting time strings to seconds for the Train dataframe

Train['Placement_-_Time']= pd.to_timedelta(Train['Placement_-_Time']).dt.total_seconds()
Train['Confirmation_-_Time']= pd.to_timedelta(Train['Confirmation_-_Time']).dt.total_seconds()
Train['Arrival_at_Pickup_-_Time']= pd.to_timedelta(Train['Arrival_at_Pickup_-_Time']).dt.total_seconds()
Train['Pickup_-_Time']= pd.to_timedelta(Train['Pickup_-_Time']).dt.total_seconds()
Train['Arrival_at_Destination_-_Time']= pd.to_timedelta(Train['Arrival_at_Destination_-_Time']).dt.total_seconds()
Train.head()
```

```python
#Converting time strings to seconds for the test_set dataframe

Test['Placement_-_Time']= pd.to_timedelta(Test['Placement_-_Time']).dt.total_seconds()
Test['Confirmation_-_Time']= pd.to_timedelta(Test['Confirmation_-_Time']).dt.total_seconds()
Test['Arrival_at_Pickup_-_Time']= pd.to_timedelta(Test['Arrival_at_Pickup_-_Time']).dt.total_seconds()
Test['Pickup_-_Time']= pd.to_timedelta(Test['Pickup_-_Time']).dt.total_seconds()
Test.head()
```

```python
#dropping outlier columns and that cannot be encoded 
Train.drop(['Order_No', 'User_Id', 'Rider_Id', 'Vehicle_Type', 'Arrival_at_Destination_-_Day_of_Month', 'Arrival_at_Destination_-_Weekday_(Mo_=_1)', 'Arrival_at_Destination_-_Time'], inplace= True, axis=1)
Train.head()
```

```python
#dropping outlier columns and that cannot be encoded  for Test dataframe
Test.drop(['Order_No', 'User_Id', 'Rider_Id', 'Vehicle_Type'], inplace= True, axis=1)
Test.head()
```

## Missing data

```python
#replacing missing data with average of each column for Train dataframe
Train.fillna(Train.mean(), inplace=True)
Train.head()
```

```python
#replacing missing data with average of each column for test_set dataset
Test.fillna(Test.mean(), inplace=True)
Test.head()
```
